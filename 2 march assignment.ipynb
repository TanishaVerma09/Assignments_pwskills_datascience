Matplotlib is a widely used Python library for creating static, animated, and interactive visualizations in Python. It provides a comprehensive set of tools for creating various types of plots and charts.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:
Line Plot: A line plot is used to display data points connected by straight lines. It is suitable for visualizing trends and patterns in continuous data.

Scatter Plot: A scatter plot is used to display the relationship between two variables. Each data point is represented as a dot, and the position of the dot corresponds to the values of the two variables.

Bar Plot: A bar plot (or bar chart) is used to compare categorical data. It displays data as rectangular bars, where the height of each bar represents the value of a specific category.

Histogram: A histogram is used to visualize the distribution of a continuous variable. It divides the data into bins and displays the frequency or count of data points falling within each bin.

Pie Chart: A pie chart is used to represent data as a circular graph, where the size of each "slice" corresponds to the proportion or percentage of a specific category within a whole.

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each with its own axes, enabling you to display multiple plots or charts together in a structured manner.

The subplot() function takes three arguments: nrows, ncols, and index. These arguments specify the number of rows and columns in the grid and the index of the current subplot.

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 =np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
fig.tight_layout()
plt.show()
```

Bar Plot: A bar plot (or bar chart) is used to compare categorical data. It displays data as rectangular bars, where the height of each bar represents the value of a specific category.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Name of company")
plt.ylabel("Profit of the company")
plt.title("Profits")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("Name of company")
plt.ylabel("Profit of the company")
plt.title("Profits")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. It provides a concise summary of several important statistical measures and allows for easy comparison between different groups or variables.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
