In Bokeh, glyphs are visual shapes or markers that represent data points in a plot. They are fundamental components used to display data in various ways, such as circles, squares, lines, bars, patches, etc. Glyphs are added to a Bokeh plot using specific glyph methods available in the figure object.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
plot = figure(title="Glyph Example", x_axis_label='X-axis', y_axis_label='Y-axis', width=400, height=400)
plot.circle(x, y, size=10, color='navy', alpha=0.8, legend_label='Circle Glyph')
plot.line(x, y, line_width=2, color='red', alpha=0.8, legend_label='Line Glyph')
plot.square(x, y, size=15, color='green', alpha=0.6, legend_label='Square Glyph')
x_patch = [1, 3, 2]
y_patch = [2, 3, 6]
plot.patch(x_patch, y_patch, alpha=0.4, line_width=2, color='orange', legend_label='Patch Glyph')
plot.legend.title = 'Glyphs'
plot.legend.location = 'top_left'
plot.legend.click_policy = 'hide'
output_notebook()
show(plot)
```

A Bokeh server is a Python process that allows you to create interactive web applications and dashboards with Bokeh plots. It enables you to build dynamic and responsive visualizations that can be updated in real time, either based on user interactions or through periodic updates triggered by the server.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
from bokeh.io import output_notebook, show
from bokeh.models.widgets import Button
from bokeh.events import ButtonClick
```

```python
# Prepare the initial data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a ColumnDataSource to hold the data (needed for real-time updates)
source = ColumnDataSource(data=dict(x=x, y=y))

# Create the plot
plot = figure(title="Real-time Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis', width=400, height=400)
plot.circle('x', 'y', size=10, color='navy', alpha=0.8, source=source)

# Create an update button
update_button = Button(label="Update Data", button_type="success")

# Define a function to update the data on button click
def update_data():
    # Update the x and y data with new values
    new_x = [1, 2, 3, 4, 5]
    new_y = [10, 2, 8, 6, 3]
    source.data = dict(x=new_x, y=new_y)

# Attach the update_data function to the button's on_click event
update_button.on_event(ButtonClick, update_data)
```

```python
# Create a layout for the app, including the plot and the update button
layout = column(plot, update_button)

# Create the app and add the layout to the document
curdoc().add_root(layout)

# If you are running the Bokeh server in a Jupyter Notebook, display the app inline
output_notebook()

# If you are running the Bokeh server as a standalone Python script, use show
# show(layout)
```
